```python
# import libraries

import csv
import pandas as pd
```

```python
# source 1, only sales data based on platform, genre, and publisher

url_1 = "data/csv/sales_data.csv"
sales_data = pd.read_csv(url_1)
sales_data = sales_data.dropna()
sales_data.to_csv("data/csv/sales_data_clean.csv", index= False, header = True)

sales_data
```

```python
# source : sales data+ rating and critic scores:

url_2 = "data/csv/rating_sale.csv"
rating_sale = pd.read_csv(url_2)
rating_sale = rating_sale.dropna()
rating_sale.to_csv("data/csv/rating_sale_clean.csv", index= False, header = True)
rating_sale
```

```python
# source 3: scraping whatoplay: "https://whatoplay.com/best?sort-by=gamer_score" => game score, play score, critic score

url_3 = "data/csv/scraping_result.csv"

whatoplay_data = pd.read_csv(url_3)

whatoplay_data = whatoplay_data.dropna()
whatoplay_data.to_csv("data/csv/whatoplay_data_clean.csv")

whatoplay_data
```

```python
total_data = pd.merge(rating_sale, whatoplay_data, how = "inner", on=["Name", "Platform"] )

total_data 
```

```python
# Store the final table as csv file

total_data.to_csv("data/csv/master_table_clean.csv", index= False, header = True)
```

```python
# convert csv to jason format

total_data.to_json ("data/json/master_table.json")
rating_sale.to_json ("data/json/rating_sale.json")
sales_data.to_json("data/json/sales_data.json")
whatoplay_data.to_json ("data/json/whatoplay_data.json")
```
